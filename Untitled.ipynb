```python
import os
import numpy as np
import cv2
import mrcnn.config
import mrcnn.utils
from mrcnn.model import MaskRCNN
import matplotlib.pyplot as plt
from mrcnn import visualize
```

```python
class MaskRCNNConfig(mrcnn.config.Config):
    NAME = "coco_pretrained_model_config"
    IMAGES_PER_GPU = 1
    GPU_COUNT = 1
    NUM_CLASSES = 1 + 80  # COCO dataset has 80 classes + one background class
DETECTION_MIN_CONFIDENCE = 0.6
```

```python
model=MaskRCNN(mode='inference',model_dir='./logs',config=MaskRCNNConfig())
```

```python
model.load_weights('mask_rcnn_coco.h5',by_name=True)
```

```python

def get_car_boxes(box,ids):
    
    boxes=[]
    
    for i in range(len(ids)):
        if ids[i] in [3,4,8,9]:
            boxes.append(box[i])
            
    return boxes 
```

```python
images=os.listdir('camera3')
parking_space=[]
threshold=4
len(images)
```

This model does not require training for next steps. This method works on live cctv footage from parking lot by breaking them into frames and then learning and implementing simoultaneously from them. But due to unavailibility of video, we tested this on images of same parking lot. It can work on any parking lot, just the below steps must be same. Now we will show output when one by one image is given to it.

```python

img=plt.imread('camera3/'+images[0])
img=cv2.resize(img,(1000,750))
results=model.detect([img])
boxes=get_car_boxes(results[0]['rois'],results[0]['class_ids'])
car_boxes_full=[]
    
for i in boxes:
    y0,x0,y1,x1=i
    flag=0
    for j in parking_space:
        y2,x2,y3,x3=j
        if y2-threshold<=y0<=y2+threshold and x2-threshold<=x0<=x2+threshold and y3-threshold<=y1<=y3+threshold and x3-threshold<=x1<=x3+threshold:
            car_boxes_full.append([y0,x0,y1,x1])
            flag=1
            
    if flag==0:
        parking_space.append([y0,x0,y1,x1])
            
for i in parking_space:
    y0,x0,y1,x1=i
    b=cv2.rectangle(img,(x0,y0), (x1, y1),(0,255,0),3)
    
for i in boxes:
    y0,x0,y1,x1=i
    flag=0
    for j in parking_space:
        y2,x2,y3,x3=j
        if y2-threshold<=y0<=y2+threshold and x2-threshold<=x0<=x2+threshold and y3-threshold<=y1<=y3+threshold and x3-threshold<=x1<=x3+threshold:
            car_boxes_full.append([y0,x0,y1,x1])
            flag=1    

for i in car_boxes_full:
    y0,x0,y1,x1=i
    b=cv2.rectangle(img,(x0,y0), (x1, y1),(255,0,0),3)
        

plt.imshow(img)
        
```
